## Chapter 4 Linear Regression

```python
# Import our main libraries
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
url = "http://peopleanalytics-regression-book.org/data/ugtests.csv"
ugtests = pd.read_csv(url)
```

```python
ugtests
```

```python
ugtests.describe()
```

```python
sns.pairplot(ugtests)
```

```python
pearsoncorr = ugtests.corr(method='pearson')
pearsoncorr
```

```python
sns.heatmap(pearsoncorr, xticklabels=pearsoncorr.columns,
           yticklabels=pearsoncorr.columns,
           cmap='RdBu_r',
           annot=True,
           linewidth=0.5)
```

```python
# Simple Regression
model1 = smf.ols(formula = "Final ~ Yr3 ", data = ugtests)
ugtests_model1 = model1.fit()
print(ugtests_model1.summary())
```

```python
# Multiple Regression 
model = smf.ols(formula = "Final ~ Yr3 + Yr2 + Yr1", data = ugtests)
ugtests_model = model.fit()
print(ugtests_model.summary())
```

```python
x = ugtests['Final']
y = ugtests['Yr3']

ax = sns.regplot(x, y)
```
